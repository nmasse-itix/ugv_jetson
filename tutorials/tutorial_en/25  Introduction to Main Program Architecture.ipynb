# Main Program Architecture Introduction

## File Structure and Function Overview

- ugv_rpi
    - [Folder] AccessPopup (used for network connection related functions)
    - [Folder] controllers (used for visual and audio interaction, upper-level computer information acquisition, and lower-level machine control functions)
    - [Folder] scripts (scripts for automatic installation, starting JupyterLab server, and configuring the main program and JupyterLab to run automatically on startup)
    - [Folder] templates (resources for the web application, photos and videos are also stored here)
    - [Folder] tutorial (interactive tutorials in Chinese and English) Interactive tutorial in Chinese)
    - LICENSE (GPL-3.0 open source license)
    - app.py (main program, contains web-socket and Flask related functions)
    - asound.conf (sound card configuration file)
    - config.yaml (configuration file, used to set certain parameters)
    - requirements.txt (Python project dependency libraries)
    - sunny (sunny-ngrok program)

## Installation Script

In the project folder, there is a file named `setup.sh`, written in shell script, which helps automatically configure the robot product's host PC, including setting up the serial port, configuring the camera, creating the project virtual environment, and installing dependencies. These steps are already configured in the image of the SD card provided from the factory.

Usage of the installation script: The installation process requires downloading and installing many dependencies from the internet. For regions with special network conditions, we recommend using the method of directly downloading the image file from our official website to install the product.

`setup.sh` needs to be run with root privileges.
> cd ugv_rpi/scripts/
> sudo chmod +x setup.sh
> sudo ./setup.sh

## Auto-Run Program

The `autorun.sh` file in the project folder is used to configure the product's main program (`app.py`) and JupyterLab (`start_jupyter.sh`) to auto-run at startup (as a user, not root), and it also generates the JupyterLab configuration file.

`autorun.sh` needs to be run with user privileges.
> cd ugv_rpi/scripts/
> sudo chmod +x autorun.sh
> sudo chmod +x start_jupyter.sh
> ./autorun.sh

## Updating the Project

This project will be continuously updated. When you need to update it, simply navigate to the project folder and use the `git pull` command to update. Note that this operation will overwrite any changes you have made. After updating, you need to modify the parameters in `config.yaml`, set the chassis type and module type. You can also use the command-line tool in the web control interface to configure the product type.
> cd ugv_rpi/
> git pull

After the device restarts, open the web control interface and enter the command to set the product type, for example:
> s 20

The first digit `2` represents the chassis type, and the second digit `0` represents no module or a pan-tilt module.

First digit (Chassis Type):
- 1: RaspRover
- 2: UGV Rover
- 3: UGV Beast

Second digit (Module Type):
- 0: None/PT
- 1: RoArm M2
- 2: Pan-Tilt
- 3: RoArm M3

This configuration only needs to be done once after updating or reinstalling and will be saved in `config.yaml`.
