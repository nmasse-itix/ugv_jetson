### Configuring Services to Start Automatically at Boot with systemd

Linux uses **systemd** to manage background services, including starting, stopping, restarting, and enabling programs to start at boot. You just need to create a `.service` file to make a Python program or script run automatically.

## 1. Create a User-Level Service File

User-level service files are stored at:

```
~/.config/systemd/user/
```

If the directory doesn't exist, create it:

```bash
mkdir -p ~/.config/systemd/user
```

Create a service file (for example: `ugv-app.service`):

```bash
nano ~/.config/systemd/user/ugv-app.service
```

Enter the following content:

```ini
[Unit]
Description=UGV Python App
After=sound.target pipewire.service
Wants=pipewire.service

[Service]
ExecStart=/home/ws/ugv_rpi/ugv-env/bin/python /home/ws/ugv_rpi/app.py >> /home/ws/ugv_rpi/ugv-app.log 2>&1
Restart=always
Environment=XDG_RUNTIME_DIR=/run/user/%U
WorkingDirectory=/home/ws/ugv_rpi

[Install]
WantedBy=default.target
```

## 2. Reload systemd Configuration

```bash
systemctl --user daemon-reload
```

## 3. Enable Service to Start at Boot

```bash
systemctl --user enable ugv-app.service
```

## 4. Start the Service Manually

```bash
systemctl --user start ugv-app.service
```

## 5. Check Service Status

```bash
systemctl --user status ugv-app.service
```

## 6. View Service Logs

```bash
cat /home/ws/ugv_rpi/ugv-app.log
```
