# YAML Configuration File Settings

## What is YAML?

YAML (YAML Ain't Markup Language) is a human-readable data serialization format used to represent complex data structures. Its main uses include configuration files, data exchange and storage, and passing data to programs.

## Advantages of YAML

### Human Readability

YAML uses indentation and a well-readable structure, making documents easy to read and understand. It is not as verbose as XML or JSON, and is closer to natural language.

### Ease of Writing and Editing

YAML's syntax is concise and clear, requiring no additional markup symbols (such as XML tags or JSON braces), making it easier to write and edit.

### Support for Complex Data Structures

YAML supports complex data structures such as nested structures, lists, and dictionaries, easily representing various types of data.

### Extensibility

YAML allows the use of tags and anchors to represent relationships between objects, enabling data reuse and referencing, thus improving data extensibility.

### Language Independence

YAML is a universal data serialization format that is independent of any specific programming language, and therefore can be easily parsed and generated by various programming languages.

## This Product's config.yaml

In this product's config.yaml, we configure some key robot-related parameters:

### args_config Robot Parameter Settings

- max_speed: Maximum speed

- max_turn_speed: Maximum turning speed

- slow_speed: Low speed

- max_rate: Ratio of maximum speed

- mid_rate: Ratio of mid speed

- min_rate: Ratio of low speed

### audio_config Audio-Related Settings

- audio_output: Whether to use audio output

- default_volume: Default volume

- min_time_bewteen_play: Minimum audio playback interval

- speed_rate: TTS speech rate

### base_config Basic Information

- robot_name: Product name

- main_type: Chassis type (1-RaspRover, 2-UGV Rover, 3-UGV Beast)

- module_type: Module type (0-No module, 1-Robotic arm, 2-Gimbal)

- sbc_version Host Computer Version

### OpenCV Parameter Settings

- color_list: Three colored balls for color recognition

- default_color: Default target color for color recognition

- track_base: Main body for color tracking

- line_lower: LAB LOWER value for the yellow line color

- line_upper: LAB UPPER value for the yellow line color

### sbc_config Host Computer Settings

- feedback_interval: Interval for receiving feedback information

- disabled_http_log: Disable HTTP server logging information

### Video Resolution Settings

- default_res_h: Height in pixels

- default_res_w: Width in pixels

### cmd_config Command Type Codes

These codes are related to the corresponding commands and command definitions in the lower-level program. They are used for communication and debugging between the host and lower-level machines. Changing these involves modifying the lower-level program.

### Code Function Names

These codes correspond to functions. The front-end page also needs to load this .yaml file to obtain these configurations. This ensures that different buttons correspond to different functions when the front-end web application communicates with the back-end. No changes are needed unless absolutely necessary.

### Feedback Message Codes

These codes correspond to feedback message types. Some feedback messages are from the chassis to the host computer, and some are from the back-end to the front-end. The back-end and front-end use the same .yaml file to unify these codes and feedback message types. No changes are needed unless absolutely necessary.
