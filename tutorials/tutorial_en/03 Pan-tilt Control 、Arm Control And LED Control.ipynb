# Gimbal Control, Robotic Arm Control, and LED Control
## Gimbal Control
### Basic Introduction

Products with a gimbal include two servos: a horizontal servo and a tilt servo. The horizontal servo controls the rotation of the gimbal in the horizontal plane, with a rotation range of ±180° (total range 360°); the tilt servo controls the rotation of the gimbal in the vertical plane, with a rotation range of -30° to 90° (total range 120°).

For products without a built-in gimbal, it is also possible to add a gimbal to the UGV by yourself.

```python
# Input the library for base control
from base_ctrl import BaseController

base = BaseController('/dev/ttyTHS1', 115200)
```

In the code block above, we import and instantiate the library used for chassis control. Next, we control the gimbal movement by changing the angles of the gimbal's horizontal and tilt servos.

Change the values in the following code. In this code:
- input_x is the rotation angle of the gimbal's horizontal servo. The horizontal servo PAN axis has a rotation range of ±180° (total range 360°);
- input_y is the rotation angle of the gimbal's tilt servo. The tilt servo TILT axis has a rotation range of -45° to 90° (total range 135°);
- input_speed is the speed at which the gimbal rotates. When the speed value is 0, the rotation is at its fastest;
- input_acc is the acceleration at the start and end of the gimbal's movement. The smaller the value, the smoother the start and stop. When the acceleration value is 0, the rotation is performed at the maximum acceleration.

Run the following code, and you will see the gimbal move 45° to the right and 45° upward before stopping.

```python
input_x = 45
input_y = 45
input_speed = 0
input_acc = 0

base.gimbal_ctrl(input_x, input_y, input_speed, input_acc)
```

In addition to controlling the gimbal's movement by changing the rotation angles of its two servos, you can also directly control the gimbal to move continuously.

Change the values in the following code. In this code:

- input_x is the horizontal servo rotation mode of the gimbal. When the value is -1, it means the gimbal continuously rotates to the left (clockwise); when the value is 1, it means the gimbal continuously rotates to the right (counterclockwise); when the value is 0, it means to stop rotating.
- input_y is the tilt servo rotation mode of the gimbal. When the value is -1, it means the gimbal's tilt servo continuously tilts downward (looking down); when the value is 1, it means the tilt servo continuously tilts upward (looking up); when the value is 0, it means to stop rotating.
- input_speed is the speed of the gimbal rotation.

The gimbal will automatically return to its middle position only when both input_x and input_y are set to 2.

Run the following code, and the gimbal will move to the left until it reaches 180° and then stop.

```python
input_x = -1
input_y = 0
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

Run the following code, and the gimbal will move upward and stop when it reaches 90°.

```python
input_x = 0
input_y = 1
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

Finally, run the following code to return the gimbal to the center position.

```python
input_x = 2
input_y = 2
input_speed = 0

base.gimbal_base_ctrl(input_x, input_y, input_speed)
```

```python
# You can modify the JSON parameters in the code block below according to the type of product you have to configure the chassis type:

# Chassis Type Selection

# The value of "main" represents the chassis type:
# - 1: RaspRover (four-wheel drive chassis)
# - 2: UGV Rover (six-wheel drive chassis)
# - 3: UGV Beast (tracked chassis)

# The value of "module" represents the module type:
# - 0: No module installed or the pan-tilt module can be set to this value
# - 1: M2 robotic arm module (if the robotic arm is not installed, the chassis will fail to ping and report joint errors)
# - 2: Pan-tilt module (if the pan-tilt is not installed, the chassis will fail to ping and report joint errors)
# - 3: M3 robotic arm module (if the robotic arm is not installed, the chassis will fail to ping and report joint errors)

# For M2 robotic arm
base.base_json_ctrl({"T":900,"main":2,"module":1})

# For M3 robotic arm
# base.base_json_ctrl({"T":900,"main":2,"module":3})
```

Change the values in the following code. In the code below:

- shoulder: the rotation angle of the shoulder joint servo. 
- elbow: the rotation angle of the elbow joint servo. 
- wrist: the rotation angle of the first wrist joint servo. 
- roll: the rotation angle of the second wrist joint servo. 
- hand: the rotation angle of the end joint servo.

Running the following code will allow you to see the robotic arm move from its current position to the specified initial position and then stop.

```python
base_radian = 0.0
shoulder_radian = 0.0
elbow_radian = 1.5708
wrist_radian = 0.0
roll_radian = 0.0
hand_radian = 3.1415926
spd = 0.0
acc = 30.0    

# For the m2 robotic arm
base.base_json_ctrl({"T":102,"base": base_radian,"shoulder": shoulder_radian,"elbow": elbow_radian,"hand": hand_radian,"spd": spd,"acc": acc})

# For the m3 robotic arm
# base.base_json_ctrl({"T":102,"base": base_radian,"shoulder": shoulder_radian,"elbow": elbow_radian,"wrist": wrist_radian,"roll": roll_radian,"hand": hand_radian,"spd": spd,"acc": acc})
```

## LED Control
### Basic Introduction
The WAVE ROVER and UGV series products have two integrated 12V switch control interfaces on the driver board (the actual maximum voltage will vary with the battery voltage), which are controlled via MOSFETs through the ESP32's IO4 and IO5 pins. Each pin corresponds to two interfaces, totaling four 12V switch control interfaces. By default assembly, IO4 controls the chassis headlights (WAVE ROVER does not have chassis headlights), and IO5 controls the front headlights. You can control the on/off state and voltage level of these two interfaces by sending the corresponding commands to the lower computer. However, due to inherent delays in MOSFET control, the PWM output from the ESP32's IO pins does not linearly correspond to the actual voltage output of the interfaces.

For products without LED lights, you can expand the two 12V switch control interfaces with LED lights rated for 12.6V (normally 12V-rated lights are also acceptable, as the product's UPS will not charge the battery above 12V to ensure safety and protect the battery). You can also use the remaining switch control interfaces to expand other peripherals—for example, a 12V-rated water bullet gun gearbox can be connected directly to the interface controlled by IO5, enabling automatic aiming and shooting functionality.

If you need to run the program within a code block, you can select the desired code block and press Ctrl+Enter to execute it. If you are using JupyterLab on a mobile device or tablet, you can press the triangular play button at the top to run the code block.

In the code block above, we imported and instantiated the library used to control the chassis. Next, we control the switch of the IO4 interface. The variable IO4_PWM is the PWM output of the ESP32's IO4 pin, with a range of 0-255. When this variable is 0, the switch controlled by IO4 is off; when this variable is 255, the voltage output by the switch controlled by IO4 is close to the BAT voltage of the UPS (the current voltage of the three series-connected lithium batteries inside the UPS).

Run the following code block to turn on the switch controlled by IO4 (turning on the chassis headlights). Note: The WAVE ROVE does not have chassis headlights, so running the following code block will have no effect. You need to run the next code block to turn on the headlight. The headlight is located on the camera gimbal. If the product is not equipped with a camera gimbal, there will be no headlight.

```python
IO4_PWM = 255
IO5_PWM = 0

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

Run the following code block to turn on the IO5-controlled interface switch (which lights up the gimbal headlamp). 
Note: If the product is not equipped with a camera gimbal, there will be no headlamp.

```python
IO4_PWM = 255
IO5_PWM = 255

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

If your product has LED lights, they should all be lit by now. You can run the following code block to dim the LED lights.

```python
IO4_PWM = 64
IO5_PWM = 64

base.lights_ctrl(IO4_PWM, IO5_PWM)
```

Finally, run the following code block to turn off the LED light.

```python
base.lights_ctrl(0, 0)
```
