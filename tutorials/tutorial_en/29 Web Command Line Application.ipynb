# Web Command Line Applications

To make product parameters easier to configure and to facilitate users adding custom features, we have designed a command-line parameter function. You can achieve the corresponding functions by entering commands in the web-based command-line tool. This section will introduce these functions in detail.

## Chassis

### Send JSON commands directly to the chassis. Refer to the relevant WIKI for specific commands.

- base -c {'T':1,'L':0,'R':0}
- base --cmd {'T':1,'L':0,'R':0}

### Enable displaying chassis information on the screen

- base -r on
- base --recv on

### Disable displaying chassis information on the screen

- base -r off
- base --recv off

### Default display of 10 seconds of command line/ESP-NOW information

- info

## Play audio

### TTS (Text-to-Speech, initial setup time is relatively long)

- audio -s what's up bro
- audio --say what's up bro

### Set volume (range between 0-1.0)

- audio -v 0.9
- audio --volume 0.9

### Play a specific sound Audio files (.mp3, .wav format, can play files from other folders) within the folder

- audio -p file.mp3
- audio --play_file file.mp3
- audio -p others/file.mp3

## Sending messages via ESP-NOW

### Adding a broadcast MAC address to a peer

- send -a -b
- send --add --broadcast
- send -a FF:FF:FF:FF:FF:FF
- send --add FF:FF:FF:FF:FF:FF

### Adding a specific MAC address to a peer

- send -a AA:BB:CC:DD:EE:FF
- send -add AA:BB:CC:DD:EE:FF

### Removing a broadcast MAC address from a peer

- send -rm -b
- send --remove --boardcast
- send -rm FF:FF:FF:FF:FF:FF
- send --remove FF:FF:FF:FF:FF:FF

### Remove a specific MAC address from a peer

- send -rm AA:BB:CC:DD:EE:FF
- send --remove AA:BB:CC:DD:EE:FF

### Broadcast Send (The broadcast MAC address needs to be added to the peer before first use)

- send -b what's up bro
- send --broadcast what's up bro

### Unicast Send (The target MAC address needs to be added to the peer before first use)

- send AA:BB:CC:DD:EE:FF what's up bro

### Multicast Send (The target MAC address needs to be added to the peer before first use; multiple MAC addresses can be added, but cannot include broadcast MAC addresses: FF:FF:FF:FF:FF:FF)

- send -g what's up bro
- send --group what's up bro

## OpenCV

### Set the target ball's color range

- cv -r [90,120,90] [120,255,200]
- cv -range [90,120,90] [120,255,200]

### Select ball color (default is only red, green, and blue)

- cv -s red
- cv --select red

### Set gimbal tracking parameters

#### Color tracking main body(optional [ugv,pt] )

- track -b pt
- track -base pt

### Set yellow line color range

- line -r [0,100,187] [255,255,255]
- line -range [0,100,187] [255,255,255]

## Chassis and module type settings

- s 20

> The first digit 2 represents the chassis type, and the second digit 0 represents no module.

Chassis Type:

1: RaspRover

2: UGV Rover

3: UGV Beast

Module Type:

0: None

1: RoArm M2

2: Pan-Tilt

3: RoArm M3
